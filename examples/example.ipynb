```python
### SETS ###
cur_case = 'buoyantCavity'
cases = {cur_case: examples_dir,}

eul_case = read_eul(
            folder=examples_dir,                # read fold 
            plot_props={'T':['foo',[290,310]]}, # consitent plots for T 
            name="example",                     # name for legend
            files=['T','k','U'],                # files to read
            skiplines=10)                       # read only every 10th entry

for name, fold in cases.iteritems():
    print name + ": ",
    cases[name] = read_sets(fold, {}, name="example")
    
symb = ['-', '-.' ,'--',':','-','-.','--']
```

# Set plots

```python
# PLOT TEMPLATES
plot_styles = ['wide','matrix','dcol','jet']
for ps in plot_styles:
    p = Plot(geom=[2,2],style=ps)
    p.pci_style()
    p.show()
```

```python
# LINESTYLES
for name, case in cases.iteritems():
    symb = ['-','--','-.',':']
    p = Plot(geom=[len(symb),1],leg=[]) # pass leg subplot as list 
    p.color_style()
    for i,s in enumerate(symb):
        for pos in range(1,9):
            _ = "y0.{}".format(pos)
            T_label = _ + '_T'
            POS = case[_]
            T = case[T_label]
            p.add(x=POS, y=T, subplot=i, symbol=s, name="bla")
    p.show()
```

```python
# SCATTER PLOTS
p = Plot(geom=[1,3],style='wide',leg=[0,1,2]) 
p.color_style()
for i,(time, case) in enumerate(eul_case.iteratetimes()):
    T = case['T']
    k = case['k']
    U = case['w']
    p.add(x=U, y=T, z=k, subplot=i)
p.show()
```
